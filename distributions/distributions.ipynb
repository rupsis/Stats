# Histograms

Histograms are used for grouping datapoints into "bins". These bins are user defined ranges that let us get a overview of where the data lies.
If were were to just graph all the data points on a line, they'd get all jumbled together and not be very useful. 

Lets take a look at a standard plot vs a histogram


```python

from numpy import random
import matplotlib.pyplot as plt


# create a random collection of values
data = random.randint(100, size=(500))
```

```python
# Plot the series of data
plt.figure()
plt.eventplot(data, orientation='horizontal', colors='b', linestyles="dotted")
plt.hlines(1,1,20)
plt.axis('off')
plt.show()



# Create a historgram of the data
num_bins = 5
n, bins, patches = plt.hist(data, num_bins, facecolor='blue', alpha=0.5)
plt.show()
```

As we can see, the first plot is just a bunch of noise. The second plot (the histogram) gives us a little more information about the data. However, since we're using such a small bin size, the data isn't all that useful. Lets plot another histogram with more bins. 

```python
# Create a historgram with more bin
num_bins = 50
n, bins, patches = plt.hist(data, num_bins, facecolor='blue', alpha=0.5)
plt.show()
```

With more bins
